```python
import numpy as numpy 
import pygame
import tkinter as tk
import math
import random
from tkinter import messagebox
```

```python
class Cube():
    rows=20
    w=500
    def __init__(self,pos,dir_x=1,dir_y=0,color=(255,0,0)):
        self.pos=pos
        self.dir_x=1
        self.dir_y=0
        self.color=color
    def move(self,dir_x,dir_y):
        self.dir_x=dir_x
        self.dir_y=dir_y
        self.pos=(self.pos[0]+self.dir_x,self.dir_y+self.pos[1])
    def draw_cube(self,window,eyes=False):
        dis=self.w//self.rows
        i=self.pos[0]
        j=self.pos[1]
        pygame.draw.rect(window,self.color,(i*dis+1,j*dis+1,dis-2,dis-2))
        
        
```

```python
class Snake():
    body=[]
    turns={}
    def __init__(self,color,pos):
        self.color=color
        self.pos=pos
        self.head=Cube(pos)
        self.body.append(self.head)
        self.dir_x=0
        self.dir_y=1
    def move(self):
        for event in pygame.event.get():
            if event.type==pygame.QUIT:
                pygame.quit()
            situation=pygame.key.get_pressed()
            for i in situation:
                if situation[pygame.K_LEFT]:
                    self.dir_x=-1
                    self.dir_y=0
                    self.turns[self.head.pos[:]]=[self.dir_x,self.dir_y]
                if situation[pygame.K_RIGHT]:
                    self.dir_x=1
                    self.dir_y=0
                    self.turns[self.head.pos[:]]=[self.dir_x,self.dir_y]
                if situation[pygame.K_UP]:
                    self.dir_x=0
                    self.dir_y=-1
                    self.turns[self.head.pos[:]]=[self.dir_x,self.dir_y]
                if situation[pygame.K_DOWN]:
                    self.dir_x=0
                    self.dir_y=1
                    self.turns[self.head.pos[:]]=[self.dir_x,self.dir_y]
        for i,j in enumerate(self.body):
            p=j.pos[:]
            if p in self.turns:
                turn=self.turns[p]
                j.move(turn[0],turn[1])
                if i==len(self.body)-1:
                    self.turns.pop(p)
            else:
                if j.dir_x == -1 and j.pos[0] <= 0: j.pos = (j.rows-1, j.pos[1])
                elif j.dir_x == 1 and j.pos[0] >= j.rows-1: j.pos = (0,j.pos[1])
                elif j.dir_y == 1 and j.pos[1] >= j.rows-1: j.pos = (j.pos[0], 0)
                elif j.dir_y == -1 and j.pos[1] <= 0: j.pos = (j.pos[0],j.rows-1)
                else: j.move(j.dir_x,j.dir_y) 
                
    def reset(self,pos):
        self.head = Cube(pos)
        self.body = []
        self.body.append(self.head)
        self.turns = {}
        self.dir_x = 0
        self.dir_y = 1
    def add_cube(self):
        tail = self.body[-1]
        dx, dy = tail.dir_x, tail.dir_y

        if dx == 1 and dy == 0:
            self.body.append(Cube((tail.pos[0]-1,tail.pos[1])))
        elif dx == -1 and dy == 0:
            self.body.append(Cube((tail.pos[0]+1,tail.pos[1])))
        elif dx == 0 and dy == 1:
            self.body.append(Cube((tail.pos[0],tail.pos[1]-1)))
        elif dx == 0 and dy == -1:
            self.body.append(Cube((tail.pos[0],tail.pos[1]+1)))

        self.body[-1].dir_x = dx
        self.body[-1].dir_y = dy
        
    def draw(self):
        for i,j in enumerate(self.body):
            if i==0:
                j.draw_cube(window)
            else:
                j.draw_cube(window)
        
        
    
            
        
        
        
```

```python
def randow_cube(rows,item):
    pos=item.body
    while True:
        x=random.randrange(rows)
        y=random.randrange(rows)
        if len(list(filter(lambda z:z.pos == (x,y), pos))) > 0:
            continue
        else:
            break
    return(x,y)
        
def drawgrid(width,rows,window):
    size=w//rows
    x=0
    y=0
    for i in range(rows):
        x+=size
        y+=size
        pygame.draw.line(window,(255,255,255),(x,0),(x,w))
        pygame.draw.line(window,(255,255,255),(0,y),(w,y))
def redraw(window):
    global w,rows
    window.fill((0,0,0))
    snake.draw()
    eat_cube.draw_cube(window)
    drawgrid(w,rows,window)
    pygame.display.update()
        
def text(statement):
    root = tk.Tk()
    root.attributes("-topmost", True)
    root.withdraw()
    messagebox.showinfo(statement)
    try:
        root.destroy()
    except:
        pass
    
    
```

```python
global w
w=500
h=500
global rows
rows=20
window=pygame.display.set_mode((w,h))
snake=Snake((255,0,0),(10,10))
eat_cube=Cube(randow_cube(rows,snake),color=(0,255,0))
clock=pygame.time.Clock()
game_over=False
while not game_over:
    pygame.time.delay(50)
    clock.tick(10)
    snake.move()
    if snake.body[0].pos==eat_cube.pos:
        snake.add_cube()
        eat_cube=Cube(randow_cube(rows,snake),color=(0,255,0))
    for x in range(len(snake.body)):
            if snake.body[x].pos in list(map(lambda z:z.pos,snake.body[x+1:])):
                print('Score: ', len(snake.body))
                text('You Lost!')
                snake.reset((10,10))
                break
        
    redraw(window)
    
```
